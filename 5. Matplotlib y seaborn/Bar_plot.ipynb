```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA','JAPAN','MEXICO','COLOMBIA','GERMANY']
population = [1000, 800,900,1000,300]
```

```python
plt.style.use('seaborn')
plt.bar(country,population, width=0.5, color = ['red','black','blue','green','lightblue'])
plt.title('Poblacion de paises')
plt.xlabel('Paises')
plt.ylabel('Habitantes')
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'),rotation =45)
plt.show()
```

```python
plt.barh(country, population)
plt.show()
```
