```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



iris = sns.load_dataset('iris')
iris.head()
```

```python
# scatterplot
plt.figure(figsize=(10,5))
sns.scatterplot(data = iris, x = 'sepal_length', y = 'petal_length',hue = 'species')
plt.show()
```

```python
# joinplot

sns.jointplot(data = iris, x = 'sepal_length', y = 'petal_length',hue = 'species')
plt.show()

```

```python
sns.scatterplot(data = iris, x='petal_width', y='sepal_width', hue='sepal_length')
legend = plt.legend()
legend.set_bbox_to_anchor((1.0, 1.0))
plt.show()
```

```python
sns.jointplot(data = iris, x='petal_width', y='sepal_width', palette='dark')
plt.show()
```

```python
sns.pairplot(data = iris, hue='sepal_length')
legend = plt.legend()
legend.set_bbox_to_anchor((1.0, 1.0))
```

```python
sns.boxplot(iris, x='species', y='sepal_length')
plt.show()
```
